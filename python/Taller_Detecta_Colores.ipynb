```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
imagen = cv2.imread('monalisa.png', 0)
cv2_imshow(imagen)
```

```python
img_color_sut = cv2.applyColorMap(imagen, cv2.COLORMAP_AUTUMN)
cv2_imshow(img_color_sut)
```

```python
img_color_sut = cv2.applyColorMap(imagen, cv2.COLORMAP_BONE)
cv2_imshow(img_color_sut)
```

```python
img_color_bone = cv2.applyColorMap(img_color_sut, cv2.COLORMAP_BONE)
cv2_imshow(img_color_bone)
```

```python
img_color_spring = cv2.applyColorMap(img_color_sut, cv2.COLORMAP_SPRING)
cv2_imshow(img_color_spring)
```
